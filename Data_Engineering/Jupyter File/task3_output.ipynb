```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder.appName("MergeParquet").getOrCreate()

# Load all Parquet files from HDFS
df = spark.read.parquet("hdfs://localhost:9000/user/student/sentiment_output/")

# Repartition to 1 partition to ensure single output file
df = df.repartition(1)

# Save as one .parquet file to a temporary directory
df.write.mode("overwrite").parquet("file:///home/student/merged_sentiment_output")
```

```python
from pyspark.sql import SparkSession
import pandas as pd

spark = SparkSession.builder.appName("SafeMergedParquetViewer").getOrCreate()

try:
    merged_df = spark.read.parquet("file:///home/student/merged_sentiment_output")
    print(" Parquet file loaded successfully.")
except Exception as e:
    print(" Error loading Parquet file:", e)
    spark.stop()

try:
    merged_df.printSchema()
except Exception as e:
    print(" Error printing schema:", e)

pd.set_option('display.max_columns', None)    
pd.set_option('display.max_colwidth', 100)    
pd.set_option('display.width', None)          

try:
    sample_df = merged_df.limit(10)  
    sample_pd = sample_df.toPandas()
    display(sample_pd)  
except Exception as e:
    print(" Error converting to Pandas:", e)
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("CountSentimentOutput") \
    .getOrCreate()

df = spark.read.parquet("file:///home/student/merged_sentiment_output")

count = df.count()

print(f" Total number of rows in sentiment_output: {count}")
```
