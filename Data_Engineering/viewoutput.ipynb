```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ReadParquetFromHDFS") \
    .getOrCreate()
```

```python
df = spark.read.parquet("hdfs://localhost:9000/user/student/sentiment_merged.parquet")
```

```python
import pandas as pd

# Show full text without truncation
pd.set_option('display.max_colwidth', None)

# Now show the DataFrame
df_pd = df.toPandas()
df_pd
```

```python
# 3. Count the number of rows
count = df.count()

# 4. Show result
print(f" Total number of rows in sentiment_output: {count}")
```

```python
unique_articles = df.select("title", "url").dropDuplicates()
print(f" Total unique articles: {unique_articles.count()}")
```

```python
from pyspark.sql.functions import sha2, concat_ws

df = df.withColumn("hash", sha2(concat_ws("", "title", "url", "full_text"), 256))
```

```python
# Total number of rows
total_rows = df.count()

# Total number of distinct rows
unique_rows = df.distinct().count()

# Count of duplicate rows
duplicate_count = total_rows - unique_rows

print(f" Total rows: {total_rows}")
print(f" Unique rows: {unique_rows}")
print(f" Duplicate rows: {duplicate_count}")
```
