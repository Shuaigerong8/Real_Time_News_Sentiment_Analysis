```python
#Nelson Tay Kai Rong
import pandas as pd
import matplotlib.pyplot as plt
from bson import ObjectId
from task4_MongoDBClient import MongoDBClient
from task4_MongoDBQueryHandler import DataQuery
from task4_ReportGenerator import ReportGenerator
from QueryOutputStyler import (
    styled_location_article_count,
    styled_keyword_sentiment_summary,
    styled_department_sentiment_count,
    styled_top_lda_topics
)

# Mapping for selective column display
QUERY_DISPLAY_COLUMNS = {
    "by_sentiment": ["title", "publishDate", "newsCompany", "predictedLabel"],
    "by_company": ["title", "newsCompany", "publishDate", "predictedLabel"],
    "by_keyword": ["title", "full_text", "publishDate", "predictedLabel"],
    "by_topic_lda": ["title", "topic_with_lda", "predictedLabel"],
    "by_topic_summary_keyword": ["title", "topic_summary", "predictedLabel"],
    "fuzzy_search_in_title": ["title", "publishDate", "newsCompany"],
    "by_date_range": ["title", "publishDate", "newsCompany", "predictedLabel"]
}

def display_selected_columns(df, query_type):
    cols = QUERY_DISPLAY_COLUMNS.get(query_type, ["title"])
    return df[cols].head(10)

def main():
    parquet_path = 'sentiment_merged.parquet'
    df = pd.read_parquet(parquet_path)
    df['_id'] = [ObjectId() for _ in range(len(df))]

    article_cols = ['_id', 'title', 'url', 'publishDate', 'newsCompany', 'full_text', 'sentence']
    enrichment_cols = ['_id', 'topic_summary', 'topic_with_lda', 'topic_with_ner', 'department', 'category', 'locations']
    model_cols = ['_id', 'predictedLabel']

    articles_df = df[article_cols].copy()
    enrichment_df = df[enrichment_cols].copy()
    model_df = df[model_cols].copy()

    mongo = MongoDBClient(
        uri="mongodb+srv://Nelson:040902010387@cluster0.eooze.mongodb.net/?retryWrites=true&w=majority",
        db_name="sentiment_analysis"
    )
    mongo.delete_all_documents()
    mongo.insert_split_documents(df.to_dict(orient='records'))

    qh = DataQuery(articles_df, enrichment_df, model_df)

    while True:
        print("\nChoose a query type:")
        print("1) By Sentiment")
        print("2) By News Company")
        print("3) By Keyword in Full Text")
        print("4) By Topic with LDA")
        print("5) By Topic Summary Keyword")
        print("6) Fuzzy Match in Title")
        print("7) Keyword Trend by Category")
        print("8) Sentiment Distribution by Company")
        print("9) Top Keywords by Sentiment")
        print("10) Company Sentiment Ratio")
        print("11) Location Article Count")
        print("12) Keyword Sentiment Summary")
        print("13) Department-wise Sentiment Count")
        print("14) Top LDA Topics")
        print("15) Articles in Date Range")
        print("0) Exit")

        choice = input("Enter choice (0-15): ").strip()

        if choice == "0":
            print("👋 Exiting.")
            break

        try:
            if choice == "1":
                val = input("Sentiment (positive/neutral/negative): ")
                df_result = qh.by_sentiment(val)
                print(display_selected_columns(df_result, "by_sentiment"))

            elif choice == "2":
                val = input("News company: ")
                df_result = qh.by_company(val)
                print(display_selected_columns(df_result, "by_company"))

            elif choice == "3":
                val = input("Keyword in full_text: ")
                df_result = qh.by_keyword(val)
                print(display_selected_columns(df_result, "by_keyword"))

            elif choice == "4":
                val = input("LDA topic: ")
                df_result = qh.by_topic_lda(val)
                print(display_selected_columns(df_result, "by_topic_lda"))

            elif choice == "5":
                val = input("Keyword in topic summary: ")
                df_result = qh.by_topic_summary_keyword(val)
                print(display_selected_columns(df_result, "by_topic_summary_keyword"))

            elif choice == "6":
                val = input("Partial title keyword: ")
                df_result = qh.fuzzy_search_in_title(val)
                print(display_selected_columns(df_result, "fuzzy_search_in_title"))

            elif choice == "7":
                keyword = input("Keyword: ")
                print(qh.keyword_frequency_by_category(keyword).to_frame("count"))

            elif choice == "8":
                print(qh.sentiment_distribution_by_company())

            elif choice == "9":
                sentiment = input("Sentiment: ")
                print(qh.top_keywords_by_sentiment(sentiment))

            elif choice == "10":
                company = input("Company: ")
                print(qh.company_sentiment_ratio(company))

            elif choice == "11":
                styled_location_article_count(qh.count_articles_by_location())
            
            elif choice == "12":
                keyword = input("Keyword: ")
                styled_keyword_sentiment_summary(qh.keyword_sentiment_summary(keyword), keyword)
            
            elif choice == "13":
                styled_department_sentiment_count(qh.department_wise_sentiment_count())
            
            elif choice == "14":
                styled_top_lda_topics(qh.top_lda_topics())

            elif choice == "15":
                start = input("Start date (YYYY-MM-DD): ")
                end = input("End date (YYYY-MM-DD): ")
                df_result = qh.by_date_range(start, end)
                print(display_selected_columns(df_result, "by_date_range"))

            else:
                print("❌ Invalid choice.")

        except Exception as e:
            print(f"❌ Error: {e}")

    # === Final Reports (Basic + Advanced) ===
    rg = ReportGenerator(articles_df, enrichment_df, model_df)
    print("\n--- Generating Final Reports ---")
    rg.report_sentiment_distribution()
    rg.report_count_by_company()
    rg.report_topics_ner_distribution()
    rg.report_category_distribution()
    
    # Advanced Reports
    rg.report_top_companies_by_sentiment("positive")
    rg.report_category_sentiment_pivot()
    rg.report_monthly_article_count()
    rg.report_topic_distribution_by_company()

if __name__ == "__main__":
    main()
```
